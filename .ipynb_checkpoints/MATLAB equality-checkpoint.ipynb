# MATLAB equality in Python



````python
# load dependencies
from matplotlib import pyplot as plt
import numpy as np
````

## Triangle function

``` MATLAB
x = linspace(0,2*pi,1000);    % also: x = 0:0.01:2*pi
y = sin(x);
z = cos(x);
hold on;        % preserve old graphic when drawing new
plot(x,y,'b');
plot(x,z,'g');
xlabel 'X values';
ylabel 'Y values';
title 'Sample Plot';
legend ('Y data','Z data');
hold off;
```

````python
x = np.linspace(0, 2*np.pi, 1000)
y, z = np.sin(x), np.cos(x)

plt.plot(x, y,'b')
plt.plot(x, z,'g')
plt.xlim(0)
plt.ylim(-1, 1)
plt.xlabel("X val")
plt.ylabel("Y val")
plt.title('Sample Plot')
plt.legend(('Y data', 'Z data'))
plt.grid()
plt.show()
````

## Piecewise function

``` matlab
t1 = linspace(0, 1, 1000);
t2 = linspace(1, 6, 1000);
y1 = t1;
y2 = 1 ./ t2;
t = [t1, t2];
y = [y1, y2];
% hold on
plot(t, y);
xlabel 't', ylabel 'y';
grid on
```

Another way: 

``` matlab
t = linspace(0, 6, 1000);
y = linspace(0, 6, 1000);
y(t > 1) = 1 ./ t(t > 1);
% hold on
plot(t, y);
xlabel 't', ylabel 'y';
grid on
```

````python
x1, x2 = np.linspace(0, 1, 1000), np.linspace(1, 6, 1000)
y1, y2 = x1, 1 / x2
x = np.concatenate((x1, x2))
y = np.concatenate((y1, y2))

plt.plot(x, y)
plt.grid()
plt.show()
````

````python
# more pythonic

x, y = np.linspace(0, 6, 1000), np.linspace(0, 6, 1000)
y[x > 1] = 1 / x[x > 1]    # towards itself

plt.plot(x, y)
plt.grid()
plt.show()
````
